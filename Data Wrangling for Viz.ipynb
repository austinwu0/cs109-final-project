```python
#for aggregating based on metric
import numpy as np
import json

def agg_metric(metric):
    gb = db[[metric, "GEOID"]].groupby("GEOID")
    medians = gb.agg([np.median])
    medians.head()
    
    #processing into json format
    lst = []
    for i in medians.iterrows():
        dct = {}
        dct["id"] = i[0]
        if np.isnan(i[1][0]):
            dct["value"] = None
        else:
            dct["value"] = i[1][0]
        lst.append(dct)
    
    for i in lst:
        if i["id"][0] == "0":
            i["id"] = i["id"][1:]
    
    import json
    fd=open("data/"+metric+".json","w")
    json.dump(lst, fd)
    fd.close()

    return medians, lst
```
