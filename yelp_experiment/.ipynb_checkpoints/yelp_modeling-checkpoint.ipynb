```python
df = pd.read_pickle('newexport.pkl')
df = df.fillna(0)
```

```python
# with help from http://planspace.org/20150423-forward_selection_with_statsmodels/
import statsmodels.formula.api as smf

def forward_selected(data, response):
    """Linear model designed by forward selection.

    Parameters:
    -----------
    data : pandas DataFrame with all possible predictors and response

    response: string, name of response column in data

    Returns:
    --------
    model: an "optimal" fitted statsmodels linear model
           with an intercept
           selected by forward selection
           evaluated by adjusted R-squared
    """
    
    remaining = set(data.columns)
    remaining.remove(response)
    selected = []
    current_score, best_new_score = 0.0, 0.0
    while remaining and current_score == best_new_score:
        scores_with_candidates = []
        for candidate in remaining:
            formula = "{} ~ {} + 1".format(response, ' + '.join(selected + [candidate]))
            score = smf.ols(formula, data).fit().rsquared_adj
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates.pop()
        if current_score < (best_new_score * 1):
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score
    formula = "{} ~ {} + 1".format(response, ' + '.join(selected))
    
    model = smf.ols(formula, data).fit()
    return model
```

```python
response=pd.read_pickle('ZipcodeMVPSnew.pkl')[['RegionName','2000_agg','2010_agg']]
response=response.rename(columns={'RegionName': 'zipcode'})
response['2010_agg_adj'] = response['2010_agg']/1.26
response['pct_delta'] = (response['2010_agg_adj']/response['2000_agg']) - 1
response.head()
```

```python
response2=pd.read_pickle('housing_rental_ratio.pkl')
response2=response2.rename(columns={'RegionName':'zipcode'})
response2['zipcode']=response2['zipcode'].astype(str)
response2['zipcode'] = [(5-len(zipcode))*'0' + zipcode if len(zipcode) < 5 else zipcode for zipcode in response2['zipcode']]
response2['ratio_pct_delta'] = response2['ratio_pct_delta'] - 1
response2.head()
```

```python
features = all_features.columns[1:]
if len(features)%3 == 0 :
    nROW = len(features)/3
else :
    nROW = np.ceil(len(features)/3.).astype(int)
nCOL = 3

f, ax = plt.subplots(nrows=nROW, ncols=nCOL, figsize=(35,70))

for i in xrange(len(features)):
    plt.subplot(nROW,nCOL,i)
    ax = all_features[features[i]].hist(bins = 30)
    ax.set_title(features[i])
```

```python
left_skewed = ['hotelstravel', 'massmedia', 'nightlife', 'professional', 'publicservicesgovt', 'realestate']
binary = ['auto', 'financialservices', 'pets', 'religiousorgs']
all_columns = df.columns.tolist()

trans_df = df.copy()

for column in left_skewed :
    trans_df[column] = [np.log(value) if value >= np.exp(-10) else -10 for value in df[column]]
    
for column in binary :
    trans_df['b_'+str(column)] = [True if value > 0 else False for value in df[column]]

trans_df = trans_df.fillna(0)
trans_df.head()
```

```python
features = trans_df.drop('zipcode',1).columns
if len(features)%3 == 0 :
    nROW = len(features)/3
else :
    nROW = np.ceil(len(features)/3.).astype(int)
nCOL = 3

fig, axes = plt.subplots(nrows=nROW, ncols=nCOL, figsize=(15,40), tight_layout=True)

for ax, feature in zip(axes.ravel(), features) :
    sns.kdeplot(trans_df[feature], shade=True, ax=ax, bw=.01)
    ax.set_title(feature)
```

```python
from sklearn.preprocessing import StandardScaler 
norm_df = trans_df.copy()

for feature in features :
    if 'b_' not in feature :
        standardizer = StandardScaler().fit(norm_df[feature])
        norm_df[feature] = standardizer.transform(trans_df[feature]).tolist()

norm_df.head()
```

```python
features = norm_df.drop('zipcode',1).columns
if len(features)%3 == 0 :
    nROW = len(features)/3
else :
    nROW = np.ceil(len(features)/3.).astype(int)
nCOL = 3

fig, axes = plt.subplots(nrows=nROW, ncols=nCOL, figsize=(15,40), tight_layout=True)

for ax, feature in zip(axes.ravel(), features) :
    sns.kdeplot(norm_df[feature], shade=True, ax=ax, bw=.01)
    ax.set_title(feature)
```

```python
from sklearn.cross_validation import train_test_split
# split into training and test sets
itrain, itest = train_test_split(xrange(trans_df.shape[0]), train_size=0.8)

mask=np.ones(trans_df.shape[0], dtype='int')
mask[itrain]=1
mask[itest]=0
mask = (mask==1)

print 'samples in training set = %s \nsamples in test set = %s'%(mask.sum(), mask.shape[0] - mask.sum())
```

```python
train_norm_df = norm_df[mask]
test_norm_df = norm_df[~mask]
train_norm_df.shape, test_norm_df.shape
```

```python
train_ratio_2015 = pd.merge(train_norm_df, response2[['zipcode','ratio_2015']], on='zipcode')
test_ratio_2015 = pd.merge(test_norm_df, response2[['zipcode','ratio_2015']], on='zipcode')
train_ratio_2015_model = forward_selected(train_ratio_2015.drop('zipcode',1), 'ratio_2015')
ratio_2015_model_features = train_ratio_2015_model.params.index.values.tolist()
train_ratio_2015_model.summary()
```

```python
ratio_2015_model_features = [el.replace('[T.True]','') for el in ratio_2015_model_features if el != 'Intercept'] + ['ratio_2015']
ratio_2015_model_features
```

```python
train_ratio_delta = pd.merge(train_norm_df, response2[['zipcode','ratio_pct_delta']], on='zipcode')
test_ratio_delta = pd.merge(test_norm_df, response2[['zipcode','ratio_pct_delta']], on='zipcode')
train_ratio_delta_model = forward_selected(train_ratio_delta.drop('zipcode',1), 'ratio_pct_delta')
ratio_delta_model_features = train_ratio_delta_model.params.index.values.tolist()
train_ratio_delta_model.summary()
```

```python
ratio_delta_model_features = [el.replace('[T.True]','') for el in ratio_delta_model_features if el != 'Intercept'] + ['ratio_pct_delta']
ratio_delta_model_features
```
